1. Impoting Dependencies :

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

2. Data Collcetion and Analysis:

A- Loading dataset of Boston House Price:

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
col_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
data = pd.read_csv(url, delim_whitespace=True, names=col_names)
```

B- View the data (head)

```python
data.head()
```

C-Number of row & columns:

```python
data.shape
```

D- Check for missing values;

```python
data.isnull().sum()
#We don't have any missing valie
```

2. Statisctical measures :

A. General Statistic:

```python
data.describe()
```

3. Correlation between various features in the dataset:

```python
correlation = data.corr()
```

A. Constructing a heatmap to understand the correlation

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, annot=True, annot_kws={'size':8}, cmap='Blues')
```

4. Spilitti,g the data and Target (Price):

A- Separating the data and labels

```python
X = data.drop(['MEDV'], axis= 1)
Y = data['MEDV']
print(X)
print(Y)
```

B- Spiliting the data into Training & test data:

```python
X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape,X_train.shape, X_test.shape)
```

5. Model Training:

A- Loading the model

```python
model = XGBRegressor()
```

B- Training the model:

```python
model.fit(X_train,Y_train)
```

6. Evaluation of model:

A. Prediction on training data:

```python
data_prediction = model.predict(X_train)
print(data_prediction)
```

B- R squared erreor

```python
score_1 = metrics.r2_score(Y_train, data_prediction)
```

C- Mean Absolute Error

```python
score_2 = metrics.mean_absolute_error(Y_train, data_prediction)
```

```python
print("R quared error :", score_1)
print("Mean Absolute Error :", score_2)
```

D- Visualisation the actual Prices and Predict prices:

```python
plt.scatter(Y_train, data_prediction)
plt.xlabel('Actual prices')
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

E. Prediction on testing data:

```python
data_prediction_testing= model.predict(X_test)
print(data_prediction_testing)
```

```python
score_1test = metrics.r2_score(Y_test, data_prediction_testing)
score_2test = metrics.mean_absolute_error(Y_test, data_prediction_testing)
print("R quared error :", score_1test)
print("Mean Absolute Error :", score_2test)
```
